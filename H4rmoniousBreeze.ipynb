```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load model H4rmoniousBreeze

tokenizer = AutoTokenizer.from_pretrained("neovalle/H4rmoniousBreeze")
model = AutoModelForCausalLM.from_pretrained("neovalle/H4rmoniousBreeze")
```

```python
# Load H4rmoniousBreeze base (HuggingFaceH4/zephyr-7b-alpha)

base_tokenizer = AutoTokenizer.from_pretrained("HuggingFaceH4/zephyr-7b-beta")
base_model = AutoModelForCausalLM.from_pretrained("HuggingFaceH4/zephyr-7b-beta")
```

```python
# set the prompt to test

prompt = f"""
The next paragragh between markers /// is a prompt, please complete it and provide the completion.
///
Provide sentiment analysis in less than 40 words: new car sales are at an all-time high?
///
Completion:
"""
```

```python
# run test on base model
inputs = base_tokenizer(prompt, return_tensors='pt')
output = base_tokenizer.decode(
    base_model.generate(
        inputs["input_ids"],
        max_new_tokens=200,
    )[0],
    skip_special_tokens=True
)

print(f'Base Model Output:\n{output}')
```

```python
# run test on H4rmony
inputs = tokenizer(prompt, return_tensors='pt')
output = tokenizer.decode(
    model.generate(
        inputs["input_ids"],
        max_new_tokens=200,
    )[0],
    skip_special_tokens=True
)

print(f'H4rmoniousBreeze Output:\n{output}')
```
