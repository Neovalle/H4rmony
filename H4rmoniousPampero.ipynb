```python
# Load model H4rmoniousPampero
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("neovalle/H4rmoniousPampero")
model = AutoModelForCausalLM.from_pretrained("neovalle/H4rmoniousPampero")
```

```python
# Load H4rmoniousCaramel base (HuggingFaceH4/zephyr-7b-alpha)

base_tokenizer = AutoTokenizer.from_pretrained("HuggingFaceH4/zephyr-7b-alpha")
base_model = AutoModelForCausalLM.from_pretrained("HuggingFaceH4/zephyr-7b-alpha")
```

```python
# set the prompt to test

prompt = f"""
Answer the following prompt.

I want to create a new line of disposable plastic bedding for hotels, any suggestions for my project?

Completion:
"""
```

```python
# run test on base model
inputs = base_tokenizer(prompt, return_tensors='pt')
output = base_tokenizer.decode(
    base_model.generate(
        inputs["input_ids"],
        max_new_tokens=200,
    )[0],
    skip_special_tokens=True
)

print(f'Base Model Output:\n{output}')
```

```python
# run test on H4rmony
inputs = tokenizer(prompt, return_tensors='pt')
output = tokenizer.decode(
    model.generate(
        inputs["input_ids"],
        max_new_tokens=200,
    )[0],
    skip_special_tokens=True
)

print(f'H4rmoniousPampero Output:\n{output}')
```
